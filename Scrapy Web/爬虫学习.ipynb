```python

def get_html(url):
    header={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:67.0) Gecko/20100101 Firefox/67.0'}
    try:
        response=requests.get(url,headers=header)
        # 判断网页是否正确返回
        if response.status_code==200:
            return response.content.decode('utf-8')
        else:
            print("{0}网页请求状态码错误!{0}".format("-"*30))
    except Exception as e:
        print("{0}请求参数出现错误:{1}{0}".format("-"*30,e))



```

```python

def parse_html(html):
    # 通过xpath提取
    html=etree.HTML(html)
    goods=html.xpath('/html/body/div/div/div/div/div[1]/div[1]/div[9]/div/div[1]/div/div/a/div/div/div[1]/div/div/div/text()')
    price=html.xpath('/html/body/div/div/div/div/div[1]/div[1]/div[9]/div/div[1]/div/div/a/div/div/div[1]/div[1]/div/text()')

    # print(index,title,actor,releasetime,score1,score2)
    tlist=[]
    for i in range(len(goods)):
        tlist.append([goods[i],price[i]])
    for line in tlist:
        print(line)
    return tlist


```

```python
def save_to_csv(tlist):
    with open("虾皮goods.csv",'a',encoding='utf-8-sig')as f:
       for line in tlist:
           for word in line:
               f.write(word+',')
           f.write('\n')

```
