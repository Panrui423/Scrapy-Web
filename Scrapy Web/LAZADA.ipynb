```python
next_params = params
# 创建一个空的 DataFrame
df_all = pd.DataFrame(columns=['name','price', 'brandId', 'brandName', 'itemId', 'inStock','image'])
j=0
while next_params:
    response = requests.get('https://www.lazada.co.id/beli-makanan-minuman-hasil-segar/', params=next_params, cookies=cookies, headers=headers)
    response.status_code
    data = json.loads(response.text)
    if 'mods' in data:
        length_data = len(data['mods']['listItems'])
    else:
        break
        
    name = {}
    brandId = {}
    price = {}
    brandName = {}
    itemId = {}
    inStock = {}
    image = {}
    for i in range(length_data):
        name[i] = data['mods']['listItems'][i]['name']
        price[i] = data['mods']['listItems'][i]['price']
        brandId[i] = data['mods']['listItems'][i]['brandId']
        brandName[i] = data['mods']['listItems'][i]['brandName']
        itemId[i] = data['mods']['listItems'][i]['itemId']
        inStock[i] = data['mods']['listItems'][i]['inStock']
        image[i] = data['mods']['listItems'][i]['image']
        
    name = pd.DataFrame.from_dict(name,orient='index')
    brandId = pd.DataFrame.from_dict(brandId,orient='index')
    price = pd.DataFrame.from_dict(price,orient='index')
    brandName = pd.DataFrame.from_dict(brandName,orient='index')
    itemId = pd.DataFrame.from_dict(itemId,orient='index')   
    inStock = pd.DataFrame.from_dict(inStock,orient='index') 
    image = pd.DataFrame.from_dict(image,orient='index') 
                       
    frames = [name, price, brandId, brandName, itemId, inStock, image]
    df = pd.concat(frames,axis = 1)
    df.columns = ['name','price', 'brandId', 'brandName', 'itemId', 'inStock','image']
    df.index = list(range(0 + j*80, 80+j*80, 1))
    df_f = [df_all,df]
    df_all = pd.concat(df_f)
    next_params['page'] = str(j+1)
    j = j+1

```
