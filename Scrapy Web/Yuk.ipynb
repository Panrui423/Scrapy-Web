```python



next_params = params
# 创建一个空的 DataFrame
df_all = pd.DataFrame(columns=['modelId', 'name', 'normalPriceLabel', 'minResellerPriceLabel','maxResellerPriceLabel','customerPrice', 'brandName','profileImageUrl'])

j=0
for cate in subId:
    next_params['subId'] = cate
    for p in range(len(typeId)):  
        typelist = typeId[p]
        for q in range(len(typelist)):
            next_params['typeId'] = typelist[q]  
            j=0
            next_params['page'] = str(1)
            while next_params:
                response = requests.get('https://evermos.com/produk/evm/v2/model/list', params=next_params, headers=headers, cookies=cookies)
                data = response.json()
                if 'data' in data and len(data['data']) > 0:
                    length_data = len(data['data'])
                else:
                    break
                modelId = {}
                name = {}
                normalPriceLabel = {}
                minResellerPriceLabel = {}
                maxResellerPriceLabel = {}
                customerPrice = {}
                profileImageUrl = {}
                brandName = {}

                for i in range(length_data):
                    modelId[i] = data['data'][i]['modelId']
                    name[i] = data['data'][i]['name']
                    normalPriceLabel[i] = data['data'][i]['normalPriceLabel']
                    minResellerPriceLabel[i] = data['data'][i]['minResellerPriceLabel']
                    maxResellerPriceLabel[i] = data['data'][i]['maxResellerPriceLabel']
                    customerPrice[i] = data['data'][i]['customerPrice']
                    profileImageUrl[i] = data['data'][i]['profileImageUrl']
                    brandName[i] = data['data'][i]['brandName']

                modelId = pd.DataFrame.from_dict(modelId,orient='index')
                name = pd.DataFrame.from_dict(name,orient='index')
                normalPriceLabel = pd.DataFrame.from_dict(normalPriceLabel,orient='index')
                minResellerPriceLabel = pd.DataFrame.from_dict(minResellerPriceLabel,orient='index')
                maxResellerPriceLabel = pd.DataFrame.from_dict(maxResellerPriceLabel,orient='index')
                customerPrice = pd.DataFrame.from_dict(customerPrice,orient='index')   
                profileImageUrl = pd.DataFrame.from_dict(profileImageUrl,orient='index') 
                brandName = pd.DataFrame.from_dict(brandName,orient='index')
                
                # 拼接数据框
                frames = [modelId, name, normalPriceLabel, minResellerPriceLabel, maxResellerPriceLabel,customerPrice, brandName, profileImageUrl]
                df = pd.concat(frames,axis = 1) # 创建空数据框
                df.columns = ['modelId', 'name', 'normalPriceLabel', 'minResellerPriceLabel','maxResellerPriceLabel','customerPrice', 'brandName','profileImageUrl']
                df_all_shape = df_all.shape[0]
                df.index = list(range(df_all_shape, df_all_shape+df.shape[0], 1))
                df_f = [df_all,df]
                df_all = pd.concat(df_f)
                next_params['page'] = str(j+2) # page参数迭代循环翻页
                j = j+1


```
