```python

import requests
import re
import json
from bs4 import BeautifulSoup
import pandas as pd

from random import randint
 
# Kecantikan
params = {
    'pageSize': '21',
    'pageIndex': '0',
    'categoryId': '280',
}
```

```python
IP_LIST = pd.read_table('ip.txt',header=None,delim_whitespace=True)[0]
```

```python
from fake_useragent import UserAgent
ua = UserAgent()
ua.chrome
```

```python
# 
category = ['163','164','166','167']
```

```python
next_params = params
# 创建一个空的 DataFrame
df_all = pd.DataFrame(columns=['product_name','product_url_image', 'product_price','product_discount_price', 'product_stock', 'product_volume', 'product_weight'])


for cate in category:
    j=0
    next_params['categoryId'] = str(cate)
    next_params['pageIndex'] = str(0)
    while next_params:
        random_ip = IP_LIST[j]
        # proxies = {'协议': '协议://IP:端口号'}
        proxie = "https://%s"%(random_ip)
        proxies = {'http': proxie}
        headers = {
        'User-Agent':ua.chrome,
        }
        response = requests.get('https://api.astronauts.id/api/product/location/0',proxies = proxies, params=next_params, headers=headers, verify=False)
        data = response.json()
        if 'content' in data and len(data['content']) > 0:
            length_data = len(data['content'])
        else:
            break
        product_name = {}
        product_url_image = {}
        product_price = {}
        product_discount_price = {}
        product_stock = {}
        product_volume = {}
        product_weight = {}

        for i in range(length_data):
            product_name[i] = data['content'][i]['product_name']
            product_url_image[i] = data['content'][i]['product_url_image']
            product_price[i] = data['content'][i]['product_price']
            product_stock[i] = data['content'][i]['product_stock']
            product_volume[i] = data['content'][i]['product_volume']
            product_weight[i] = data['content'][i]['product_weight']
            product_discount_price[i] = data['content'][i]['product_discount_price']

        product_name = pd.DataFrame.from_dict(product_name,orient='index')
        product_url_image = pd.DataFrame.from_dict(product_url_image,orient='index')
        product_price = pd.DataFrame.from_dict(product_price,orient='index')
        product_discount_price = pd.DataFrame.from_dict(product_discount_price,orient='index')
        product_stock = pd.DataFrame.from_dict(product_stock,orient='index')
        product_volume = pd.DataFrame.from_dict(product_volume,orient='index')   
        product_weight = pd.DataFrame.from_dict(product_weight,orient='index') 

        frames = [product_name, product_url_image, product_price,product_discount_price, product_stock, product_volume, product_weight]
        df = pd.concat(frames,axis = 1)
        df.columns = ['product_name','product_url_image', 'product_price', 'product_discount_price','product_stock', 'product_volume', 'product_weight']
        df_all_shape = df_all.shape[0]
        df.index = list(range(df_all_shape, df_all_shape+df.shape[0], 1))
        df_f = [df_all,df]
        df_all = pd.concat(df_f)
        next_params['pageIndex'] = str(j+1)
        j = j+1

```
